## Analyzing Temp Change

Ubah tipe data menjadi "datetime", dan menambah kolom untuk data tahun dan data bulan. Tugas ini digunakan untuk menganalisis perubahan temperatur di Kota Delhi.

```python
data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
print(data.head())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x='month', y='meantemp', hue='year')
plt.show()
```

Temperatur di Kota Delhi selalu mengalami puncaknya di bulan 5-6 (Mei-Juni), kemudian akan turun kembali pada bulan ke 10 (Oktober) sampai akhir tahun.

## Forecasting

### Facebook Prophet 

Notes : The prophet model accepts time data named as ds, and labels as y.

```python
forecast_data = data.rename(columns = {"date": "ds", 
                                       "meantemp": "y"})
print(forecast_data)
```

```python
def split_data(df, train_size):
    train_size = int(len(df) * train_size)
    train = df.iloc[0:train_size]
    test = df.iloc[train_size:]
    
    return train, test
```

```python
train_data, test_data = split_data(forecast_data, 0.8)
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(train_data)
forecasts = model.make_future_dataframe(int(len(forecast_data) * 0.2))
predictions = model.predict(test_data)
plot_plotly(model, predictions)
```

```python
from sklearn.metrics import mean_squared_error

# Hitung MSE
mse = mean_squared_error(test_data['y'], predictions['yhat'])
print(f'Mean Squared Error (MSE): {mse}')
```
